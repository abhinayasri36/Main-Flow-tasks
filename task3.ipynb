```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/kaggle/input/householdtask3/householdtask3.csv')
```

```python
display(data.head(10))
```

```python
#Scatter plot with year against own
plt.scatter(data['age'],data['income'])

#Adding Title to the Plot
plt.title("Scatter Plot")

#Setting the X and Y Labels
plt.xlabel('age')
plt.ylabel('income')

#Adding the Legends
plt.show()
```

```python
#Line chart with year against own
plt.plot(data['year'])
plt.plot(data['own'])

#Adding Title to the Plot
plt.title("Line Chart")

#Setting the X and Y Labels
plt.xlabel('year')
plt.ylabel('own')

#Adding the Legends
plt.show()
```

```python
#BAR CHART
plt.bar(data['year'],data['eqv_income'])

plt.title("Bar Chart")

#Setting the X and Y Labels
plt.xlabel('year')
plt.ylabel('eqv_income')

#Adding the Legends
plt.show()
```

```python
#HISTOGRAM
plt.hist(data['eqv_exp'])

plt.title("Histogram")

#Adding the Legends
plt.show()
```
